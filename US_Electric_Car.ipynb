# Distribution of EV registrations

* The histogram shows that most states fall into lower EV registration ranges, while only a few states appear in the higher ranges.
* The distribution is right-skewed, reflecting unequal EV adoption across states.

```python
sns.histplot(ev['Registration count'], bins=10)
plt.title('Distribution of EV Registrations')
plt.xlabel('EV Registrations')
plt.ylabel('Frequency')
plt.show()
```

# EV registrations by state

* The bar chart displays the top 10 states ranked by total electric vehicle registrations.States with taller bars represent higher EV adoption.
* The visualization helps compare EV adoption levels across states at a glance.
* The Total row represents an aggregated summary and does not correspond to any individual state. To ensure accurate state-wise comparison and avoid distortion in visualization, this row was removed before plotting.

```python
ev['State'].unique
```

```python
ev = ev[ev['State'] != 'Total']
```

```python
ev['State'].unique()
```

```python
ev_state = ev.groupby('State')['Registration count'].sum()
top_states = ev_state.sort_values(ascending=False).head(10)
top_states.plot(kind='bar')
plt.title('Top 10 States by EV Registrations')
plt.xlabel('State')
plt.ylabel('EV Registrations')
plt.show()
```

# Energy generation trend over years

* The line chart shows an upward trend in energy generation across years, with higher values observed in recent years compared to earlier periods.
* The visualization highlights long-term growth rather than short-term fluctuations.

```python
energy_year = en.groupby('YEAR')['GENERATION (MEGAWATTHOURS)'].sum()

energy_year.plot()
plt.title('Energy Generation Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Energy Generated (Megawatthours)')
plt.show()
```

# 7. POLICY IMPACT ASSESSMENT

## Objective

To assess how government policies influence:

- Electric Vehicle (EV) adoption rates

- Clean and renewable energy initiatives

### Impact of Federal Policies

Federal government policies play a nationwide role in promoting EVs and clean energy.

#### Key impacts:

- Tax credits and subsidies reduce the purchase cost of EVs

- Fuel economy and emission standards encourage manufacturers to produce electric and hybrid vehicles

- Federal funding supports renewable energy projects like solar and wind

- Investments in charging infrastructure increase consumer confidence in EV usage

#### Effect:
- Higher EV adoption across multiple states
- Increased clean energy production at a national level

### Impact of State Policies

State governments implement policies based on local priorities and resources.

#### Key impacts:

- State-level incentives (rebates, registration fee reductions)

- Zero-emission vehicle mandates in some states

- Support for renewable energy through state energy programs

- Development of local charging infrastructure

#### Effect:
- EV adoption varies significantly from state to state
- States with stronger policies show higher EV registrations

### Combined Effect on EV Adoption and Clean Energy

- States with strong federal + state support show:

        - Higher EV registrations

        - Greater renewable energy generation

- Policy support encourages a shift from fossil fuels to clean energy

- Long-term policy stability helps in sustainable transportation growth
