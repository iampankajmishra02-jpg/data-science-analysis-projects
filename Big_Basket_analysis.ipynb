## Step 6: Calculate discount percent

## Why:
## Checks how much discount each product has.

```python
df_BB['sale_price_num'] = df_BB['sale_price'].astype(str).str.replace(',', '').astype(float)
df_BB['market_price_num'] = df_BB['market_price'].astype(str).str.replace(',', '').astype(float)

df_BB['discount_percent'] = ((df_BB['market_price_num'] - df_BB['sale_price_num']) 
                              / df_BB['market_price_num']) * 100

df_BB['discount_percent'].describe()
```

## Step 8: Detect and replace outliers

## Why:
## Outliers can affect analysis.
## IQR method finds outliers and replaces them with the mean.

```python
import matplotlib.pyplot as plt

df_BB.select_dtypes(include='number').plot(kind='box', figsize=(10,5))
plt.title("Box Plot of Numeric Columns")
plt.show()
```

```python
num_cols = df_BB.select_dtypes(include='number').columns

for col in num_cols:
    Q1 = df_BB[col].quantile(0.25)
    Q3 = df_BB[col].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    
    mean_value = df_BB[col].mean()
    
    df_BB.loc[(df_BB[col] < low) | (df_BB[col] > high), col] = mean_value
```

```python
df_BB.head()
```

```python
df_BB.select_dtypes(include='number').plot(kind='box', figsize=(10,5))
plt.title("Box Plot After Outlier Treatment")
plt.show()
```
