# Restaurant Features
# Restaurants offering table booking, online ordering, and delivery consistently receive better customer ratings. These features strongly influence customer satisfaction, while opentable_support does not provide any useful informa

```python
features = ['opentable_support', 'delivery', 'Online_Order', 'Table_Booking']

with_feat = []
without_feat = []

for col in features:
    # average rating with feature
    avg_with = df[df[col] == 1]['aggregate_rating'].mean()
    # average rating without feature
    avg_without = df[df[col] != 1]['aggregate_rating'].mean()

    with_feat.append(avg_with)
    without_feat.append(avg_without)

x = range(len(features))
plt.figure(figsize=(10,5))

plt.bar([i - 0.2 for i in x], with_feat, width=0.4, label='With Feature')
plt.bar([i + 0.2 for i in x], without_feat, width=0.4, label='Without Feature')

plt.xticks(x, features, rotation=45)
plt.ylabel("Average Rating")
plt.title("Average Rating With vs Without Restaurant Features")
plt.legend()
plt.tight_layout()
plt.show()
```

# Word Cloud for Reviews
* Positive Sentiment Words:-
These words come from reviews where rating ≥ 4.0:
Good, Available, Seating, Dessert, North, Chicken, Drinks, Tandoori, etc.

* Negative Sentiment Words:-
These words come from reviews where rating < 1.5:
Available, Seating, Cost, Menu, Dessert, Staff, Quality, Tandoori, Drinks, Outlet

```python
all_words = " ".join(df['highlights'].fillna("").astype(str)).split()

top_all = Counter(all_words).most_common(15)

pos_words = " ".join(df[df['aggregate_rating'] >= 4.0]['highlights'].fillna("").astype(str)).split()
neg_words = " ".join(df[df['aggregate_rating'] < 3.5]['highlights'].fillna("").astype(str)).split()

top_pos = Counter(pos_words).most_common(10)
top_neg = Counter(neg_words).most_common(10)

def plot_bar(data, title):
    labels = [w for w,c in data]
    counts = [c for w,c in data]
    plt.bar(labels, counts)
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.show()

plot_bar(top_all, "Most Common Words (All Reviews)")
plot_bar(top_pos, "Positive Sentiment Words")
plot_bar(top_neg, "Negative Sentiment Words")
```
