# Trend: Popularity by Year
* 1.Before 1950
Popularity stays very low.
Older songs in your dataset are not popular today.

* 2.1950–1970
Popularity starts rising sharply.
This matches the rise of modern music and more recorded tracks.

* 3.1970–2000
Popularity keeps increasing slowly and steadily.

* 4.2000–2020
Popularity grows the most here.
Newer songs are much more popular because they are streamed more and have better metadata.

```python
if "year" not in df.columns:
    df["year"] = pd.to_datetime(df["release_date"], errors="coerce").dt.year

# Group by year
pop_by_year = df.groupby("year")["popularity"].mean()

print(pop_by_year)

plt.plot(pop_by_year)
plt.title("Popularity by Year")
plt.xlabel("Year")
plt.ylabel("Average Popularity")
```

# correlations between different features
* This is a weak positive correlation.

* Songs with higher energy tend to be slightly more danceable, but it’s not a strong relationship.

* Many energetic songs are not very danceable, and many danceable songs are not extremely energetic.

```python
two_features = df[['energy', 'danceability']]

corr_two = two_features.corr()

print("Correlation between energy and danceability:")
print(corr_two)

# Heatmap
plt.imshow(corr_two, interpolation='nearest')
plt.title("Heatmap: Energy vs Danceability")
plt.colorbar()
plt.xticks(range(2), ['energy', 'danceability'])
plt.yticks(range(2), ['energy', 'danceability'])
plt.show()
```

# The dataset is dominated by missing genre labels, and only a few hundred tracks have meaningful genre information. The genre feature is weak and might not be very useful for analysis or modeling unless cleaned or enriched.

```python
genre_col = 'genres'  # use your detected genre column

clean_genres = df[df[genre_col] != 'Unknown'][genre_col]

main_genres = clean_genres.astype(str).apply(lambda x: x.split(",")[0].strip())

# Count top genres
top_genres = main_genres.value_counts().head(10)

print(top_genres)

# Plot
plt.figure(figsize=(6,3))
top_genres.plot.bar()
plt.title("Top Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
